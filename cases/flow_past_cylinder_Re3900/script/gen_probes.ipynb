```python
import numpy as np
yVec = np.linspace(-1, 1, 5)
xVec = [1.06, 1.54, 2.02]
numPt = 101
for x in xVec:
    for y in yVec:
        data = np.column_stack([np.full(numPt, x), np.full(numPt, y), np.linspace(-1.5, 1.5, numPt)])
        np.savetxt(f"probesX{str(x)}Y{str(y)}.csv", data, delimiter=",", header="x,y,z", comments="", fmt="%.6f")

line1 = "      {"
line2 = "        \"OutputName\" : \"probeX{:.2f}Y{:.1f}\","
line3 = "        \"ProbesCSV\" : \"./csv/Re3900/probesX{:.2f}Y{:.1f}.csv\","
line4 = "        \"ProbeStartStep\" : \"0\","
line5 = "        \"ProbeEndStep\" : \"1000001\","
line6 = "        \"ProbeIntervalStep\" : \"100\","
line7 = "        \"Variables\" : [\"density\", \"ux\", \"uz\"]"
line8 = "      },"

for x in xVec:
    for y in yVec:
        print(line1)
        print(line2.format(x, y))
        print(line3.format(x, y))
        print(line4)
        print(line5)
        print(line6)
        print(line7)
        print(line8)
```
