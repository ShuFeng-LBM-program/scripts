```python
import numpy as np
import glob
import matplotlib.pyplot as plt


plt.rcParams.update({
    "font.size" : 15,
    "font.family": "serif",
    "font.serif": ["Times New Roman"],
    "mathtext.fontset": "stix",
})
```

```python
gamma = 1.0
c_s = 1.0 / np.sqrt(3.0)
f0 = 0.05
rhoAmp = 0.01
rhoRef = 1.0

# vis = 0.0               # tau=0.5
vis = 0.033333333       # tau=0.6
# vis = 0.166666667       # tau=1.0

Lx = 201
Ly = 201

step = 240

folderVec = [

    # tau=0.5
    # "/Users/maggie/repo/LBM-Program/amrlbm/bin/acoustic_validation/planewave_xmin_srt_tau05/probes/probeX",
    # "/Users/maggie/repo/LBM-Program/amrlbm/bin/acoustic_validation/planewave_xmin_srt_tau05_3rdEq/probes/probeX",
    # "/Users/maggie/repo/LBM-Program/amrlbm/bin/acoustic_validation/planewave_xmin_srt_smag_tau05_3rdEq/probes/probeX",

    # tau=0.6
    "/Users/maggie/repo/LBM-Program/amrlbm/bin/acoustic_validation/planewave_xmin_srt_tau06/probes/probeX",

    # tau=1.0
    # "/Users/maggie/repo/LBM-Program/amrlbm/bin/acoustic_validation/planewave_xmin_srt_tau1_postCol/probes/probeX",
    # "/Users/maggie/repo/LBM-Program/amrlbm/bin/acoustic_validation/planewave_xmin_srt_tau1/probes/probeX",

]


labelVec = [
    'LBM BGK',
    # 'LBM BGK (3rd equilibrium)',
    # 'LBM BGK (3rd equilibrium, smag)',
]

colorVec = [
    # "#910000",
    # "#119100",
    "#001AFF",
    '#FF5733',
    "#B700FF",
]
lstyleVec = [
    # '--',
    # '-.',
    ':',
    '--',
    '-.',
]
```

```python

visBulk = 2.0 / 3.0 * vis
tau = vis / c_s**2 + 0.5
tau_vi = 1.0 / c_s**2 * (4.0 / 3.0 * vis + visBulk)

omega0 = 2.0 * np.pi * f0
k0 = omega0 / c_s

omega = omega0 * (1.0 - 1.0 / 8.0 * (omega0 * tau_vi)**2)
alpha_t = omega0**2 * tau_vi / 2.0
alpha_x = omega0 * tau_vi * k0 * 0.5
k = k0 / (1.0 + 3.0 / 8.0 * (omega0 * tau_vi)**2)
k_hat = k - 1j * alpha_x


print(f"c_s = {c_s:.6f}")
print(f"f0 = {f0:6f}")
print(f"vis_bulk = {visBulk:.6f}")
print(f"tau = {tau:.6f}")
print(f"tau_vi = {tau_vi:.6f}")
print(f"omega = {omega:.6f}")
print(f"alpha_x = {alpha_x:.6f}")
print(f"k0 = {k0:.6f}")
print(f"k = {k:.6f}")
```

```python
def read_col_probe(filePath, colIdx, skipHeader=2):
    with open(filePath, 'r') as f:
        vec = np.genfromtxt(f, skip_header=skipHeader, usecols=colIdx)
        vec = vec[~np.isnan(vec)]
    f.close()
    return vec
def get_probe_coord_value(filePath, coordIdx):
    with open(filePath, 'r') as f:
        header = f.readline()
        coordValue = float(header.split('(')[1].split(')')[0].split(',')[coordIdx])
    return coordValue

coordMap = { 'x': 0, 'y': 1, 'z': 2 }
coordIdx = coordMap['x']
colIdx = 2

coordVecVec = []
dataVecVec = []
for case in range(0, len(folderVec)):
    fileList = sorted(glob.glob(f"{folderVec[case]}/probe*.txt"))
    coordVec = []
    dataVec = []
    for filePath in fileList:
        coordValue = get_probe_coord_value(filePath, coordIdx)
        # coordVec.append(coordValue)
        coordVec.append(coordValue + 100)
        stepCol = read_col_probe(filePath, 0, 2)
        dataCol = read_col_probe(filePath, colIdx)
        targetStepIdx = int(np.abs(stepCol - (step + 1)).argmin())
        dataVec.append(dataCol[targetStepIdx] - rhoRef)
    coordVecVec.append(coordVec)
    dataVecVec.append(dataVec)

```

```python

x = np.linspace(0.0, 100.0, 600)
xSafe = np.where(x == 0, 1, x)
pa = 0.38 * rhoAmp * np.exp(-alpha_x * xSafe) * np.exp(1j * (2.0 * omega0 * step - k * xSafe))
```

```python
fig1, ax1 = plt.subplots(1, 1, figsize=(12, 3), facecolor='w', edgecolor='w')
ax1.plot(xSafe, np.real(pa), c='k', linestyle='-', lw=2, alpha=1.0, label='Analytical')

for case in range(0, len(folderVec)):
    ax1.plot(coordVecVec[case][1:], dataVecVec[case][1:], c=colorVec[case], marker='o', ms=6, fillstyle='none', lw=0, label=labelVec[case])

ax1.set_title(rf'$\tau={tau:.4f}$ $(\nu={vis:.4f})$ $T={int(1/f0)}$ ($step={int(step*f0)}T$)', loc='right')
ax1.legend(loc='best', frameon=False, fontsize=13)
ax1.set_xlim([0, 100])
ax1.set_ylim([-0.01, 0.01])
ax1.set_xlabel(r'$x$')
ax1.set_ylabel(r'$\rho^\prime$')
```
