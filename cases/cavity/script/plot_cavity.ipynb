```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({
    "font.size" : 12,
    "font.family": "serif",
    "font.serif": ["Times New Roman"],
    "mathtext.fontset": "stix",
})
# plt.rcParams.update({
#     "text.usetex": True,
#     "font.family": "serif",
#     "font.serif": ["Computer Modern Roman"],
# })
```

```python
def read_col_csv(filePath, colIdx, skipHeader):
    with open(filePath, 'r') as f:
        vec = np.genfromtxt(f, skip_header=skipHeader, usecols=colIdx, delimiter=',')
    f.close()
    return vec
```

```python
lidU = 0.05

#! result
vxPathVec = [
    # "../data/cavity-1000-cas-v-x.csv",
    "../data/cavity-1000-cas-v-x-200k.csv",
    "../data/cavity-1000-bgk-v-x-200k.csv",
    # "../data/cavity-400-cas-v-x.csv",
]

yuPathVec = [
    # "../data/cavity-1000-cas-y-u.csv",
    "../data/cavity-1000-cas-y-u-200k.csv",
    "../data/cavity-1000-bgk-y-u-200k.csv",
    # "../data/cavity-400-cas-y-u.csv",
]

labelVec = [
    # "LBM Cascaded (Re=1000, 100k)",
    # "LBM Cascaded (Re=1000, 200k)",
    "LBM Cascaded (Re=1000)",
    "LBM BGK (Re=1000)",
    # "LBM Cascaded (Re=400)",
]

colorVec = [
    # '#FF5733',
    "#119100",
    "#910000",
    # "#001AFF",
]


#! reference
ghiaVXPathVec = [
    "../data/ghia-1000-v-x.dat",
    # "../data/ghia-400-v-x.dat"
]

ghiaYUPathVec = [
    "../data/ghia-1000-y-u.dat",
    # "../data/ghia-400-y-u.dat",
]

ReVec = [
    1000,
    # 400
]
markerVec = [
    's',
    # '^'
]
```

```python


resultVXVec = [[], []]
resultYUVec = [[], []]

for case in range(len(vxPathVec)):
    resultV = read_col_csv(vxPathVec[case], 3, 1)
    resultVXVec[0].append(resultV / lidU)
    resultVXVec[1].append(read_col_csv(vxPathVec[case], 0, 1))
    resultYUVec[0].append(read_col_csv(yuPathVec[case], 1, 1))
    resultU = read_col_csv(yuPathVec[case], 3, 1)
    resultYUVec[1].append(resultU / lidU)

ghiaVXVec = [[], []]
ghiaYUVec = [[], []]
for case in range(len(ghiaVXPathVec)):
    ghiaV = read_col_csv(ghiaVXPathVec[case], 1, 1)
    ghiaVXVec[0].append(ghiaV)
    ghiaVXVec[1].append(read_col_csv(ghiaVXPathVec[case], 0, 1))
    ghiaYUVec[0].append(read_col_csv(ghiaYUPathVec[case], 1, 1))
    ghiaU = read_col_csv(ghiaYUPathVec[case], 0, 1)
    ghiaYUVec[1].append(ghiaU)

fig1, ax1 = plt.subplots(1, 1, figsize=(6, 5), facecolor='w', edgecolor='w')
for case in range(len(vxPathVec)):
    ax1.plot(resultVXVec[1][case], resultVXVec[0][case], c=colorVec[case], linewidth=1, label=labelVec[case])
for case in range(len(ghiaVXPathVec)):
    ax1.plot(ghiaVXVec[1][case], ghiaVXVec[0][case], linewidth=0, marker=markerVec[case], c='k', fillstyle='none', label=f'Ghia et al. (Re={ReVec[case]})')
ax1.set_xlim(0, 1)
ax1.set_ylim(-0.6, 0.6)
ax1.legend(loc='upper right', frameon=False)
ax1.set_xlabel(r"$V/U_0$")
ax1.set_ylabel(r"$x$")

fig2, ax2 = plt.subplots(1, 1, figsize=(6, 5), facecolor='w', edgecolor='w')
for case in range(len(yuPathVec)):
    ax2.plot(resultYUVec[1][case], resultYUVec[0][case], c=colorVec[case], linewidth=1, label=labelVec[case])
for case in range(len(ghiaYUPathVec)):
    ax2.plot(ghiaYUVec[1][case], ghiaYUVec[0][case], linewidth=0, marker=markerVec[case], c='k', fillstyle='none', label=f'Ghia et al. (Re={ReVec[case]})')
ax2.set_xlim(-0.6, 1)
ax2.set_ylim(0, 1)
ax2.legend(loc='lower right', frameon=False)
ax2.set_xlabel(r"$y$")
ax2.set_ylabel(r"$U/U_0$")
```
